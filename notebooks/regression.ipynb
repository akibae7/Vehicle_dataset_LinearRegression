```python
df_new=df.drop(['name','year','fuel','seller_type','transmission','owner'], axis = 1)
df_new
```

```python
hist1 = df_new.hist(column='selling_price')
hist1
hist2 = df_new.hist(column='km_driven')
hist2
```

```python
selling_price=df_new['selling_price']
km_driven=df_new['km_driven']
```

```python
price = np.array(selling_price).reshape(-1,1)
kms = np.array(km_driven).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import math
```

```python
model = LinearRegression()
model.fit(price,kms)
```

```python
wt_pred = model.predict(price)
```

```python
plt.scatter(price,kms, color="blue")
plt.plot(price,wt_pred, color="red")
plt.title("Linear Regression")
plt.xlabel("selling price")
plt.ylabel("kilometer driven")
```
